```python
import pandas as pd
import numpy as np
import NewsSentiment
from NewsSentiment import TargetSentimentClassifier
```

```python
# df = pd.read_csv('data/TBG_unique_raw.csv')
df = pd.read_csv('data/sample_1000rows.csv')
```

```python
df.head()
```

```python
# preprocess hl1 so that no empty headlines causing error
df['hl1'].fillna("No title", inplace=True)
```

```python
articlebody = [(string,) for string in df['body']]
```

```python
headlines = [tuple(string.split(' ', 2)) + ('NA',) * (3 - len(string.split(' ', 2))) for string in df['hl1']]
headlines
```

```python
len(headlines)
```

```python
tsc = TargetSentimentClassifier()

sentiments = tsc.infer(targets=headlines)
res = []

for i, result in enumerate(sentiments):
    res.append(result[0]['class_label'])
    # print("Sentiment: ", result[0]['class_label'])
```

```python
res
```

```python
df['roberta_sentiment'] = res
```

```python
total = len(res)
percentages = {x: res.count(x) / total * 100 for x in set(res)}

# Print the percentages
for category, percentage in percentages.items():
    print(f"{category}: {percentage:.2f}%")
```

```python
df.to_csv('data/sample_1000rows_Roberta.csv', index=False)
```
